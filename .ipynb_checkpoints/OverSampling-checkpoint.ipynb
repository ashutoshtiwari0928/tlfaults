```python
import pandas as pd
```

```python
df = pd.read_csv("OptimizedDataset.csv")
```

```python
df.head()
```

```python
X = df.drop(["Type of Fault","Fault Location"],axis="columns")
Y1 = df["Type of Fault"]
Y2 = df["Fault Location"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y1,test_size = 0.2,random_state=42)
```

```python
import numpy as np
unique,count = np.unique(Y_train,return_counts=True)
```

```python
dict_ = {k:v for k,v in zip(unique,count)}
dict_
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state =42)
```

```python
X_train_res,Y_train_res = sm.fit_resample(X_train,Y_train.ravel())
```

```python
unique1,count1 = np.unique(Y_train_res,return_counts=True)
```

```python
dict1_ = {k:v for k,v in zip(unique1,count1)}
```

```python
dict1_
```

```python
X_train_res.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train_res)
```

```python
X_test_scaled = scaler.fit_transform(X_test)
X_test_scaled.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
rf_model.fit(X_train_scaled,Y_train_res)
```

```python
y_pred = rf_model.predict(X_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, y_pred,labels=rf_model.classes_)
print(cm)
```

```python
from sklearn.metrics import classification_report, accuracy_score
print(classification_report(Y_test,y_pred))
```

```python
accuracy = accuracy_score(Y_test,y_pred)
accuracy
```
